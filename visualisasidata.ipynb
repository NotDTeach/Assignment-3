```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Membaca dataset
df = pd.read_csv('movie_sample_dataset.csv')
df.head(99)
```

```python
# Histogram: Distribusi Durasi Film
plt.hist(df['duration'], bins=20)
plt.xlabel('Durasi Film (menit)')
plt.ylabel('Frekuensi')
plt.title('Distribusi Durasi Film')
plt.show()
```

```python
# Scatterplot: Korelasi Antara Pendapatan dan Anggaran
plt.scatter(df['gross'], df['budget'])
plt.xlabel('Pendapatan (juta USD)')
plt.ylabel('Anggaran (juta USD)')
plt.title('Korelasi Antara Pendapatan dan Anggaran')
plt.show()
```

```python
# Histogram: score movie
plt.boxplot(df['imdb_score'])
plt.title('Score Movie')
plt.show()
```
